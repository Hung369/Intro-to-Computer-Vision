```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(r'1.jpg')
img2 = cv2.imread(r'2.jpg')
gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
def new_func(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')
```

```python
new_func(gray1)
```

```python
new_func(gray2)
```

```python
whisk = cv2.SIFT_create()
kp1 = whisk.detect(gray1,None)
kp2 = whisk.detect(gray2,None)
```

```python
result_1 = cv2.drawKeypoints(gray1,kp1,img1,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
result_2 = cv2.drawKeypoints(gray2,kp2,img2,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
new_func(result_1[:,:,::-1])
```

```python
new_func(result_2[:,:,::-1])
```

## Points Matching

```python
img1 = cv2.imread(r'1.jpg')
img2 = cv2.imread(r'2.jpg')
```

```python
sift = cv2.SIFT_create()
kpi , des1 = sift.detectAndCompute(img1,None)
kpe , des2 = sift.detectAndCompute(img2,None)
```

```python
des1.shape #(key points, length of descriptor)
```

```python
des2.shape
```

```python
bf = cv2.BFMatcher(cv2.NORM_L2,crossCheck=True)
matches = bf.match(des1,des2)
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
img3 = cv2.drawMatches(img1,kpi,img2,kpe,matches[:20],img2,flags=2)
```

```python
plt.figure(figsize=(15,20))
new_func(img3[:,:,::-1])
```
