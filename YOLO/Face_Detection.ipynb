```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv # pip install cvlib
from cvlib.object_detection import draw_bbox
```

```python
img = cv2.imread(r'p2.jpeg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
bbox, conf = cv.detect_face(img)
bbox
```

```python
conf
```

```python
labels = list()
for i in range(len(conf)):
  labels.append('person')
```

```python
img2 = draw_bbox(img, bbox, labels, conf)
plt.imshow(img2[:,:,::-1])
```
